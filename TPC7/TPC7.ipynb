```python
import matplotlib.pyplot as plt

def convInt(s):
    return int(s)

def getAluno(texto):
    textoLimpo2=texto.replace("\"","") #igual ao strip, só que o strip é só para espaços
    campos=textoLimpo2.split(",")
    aluno=[]
    aluno.append(campos[0])
    aluno.append(campos[1])
    aluno.append(campos[2])
    aluno.append(campos[3:])
    lista=[]
    for s in aluno[3]:
        s=lista.append(int(s))
    aluno[3]=lista
    return aluno

def lerDataset(fnome):
    f=open(fnome,encoding="utf-8")
    bd=[]
    f.readline()
    for linha in f:
        bd.append(getAluno(linha))
    return bd

def chaveOrd(a):
    return a[1]

def listarDataset(bd):
    bd.sort(key=chaveOrd)
    print("id|     nome    curso     média")
    print("-------------------------------")
    for a in bd:
        print(a[0] + "|" + a[1] + "|" + a[2] + "|" + str(sum(a[3])/4))
        
def consultarDataset(bd, id):
    for a in bd:
        if a[0]==id:
            print (a)
            print("Média:"+ str(sum(a[3])/4))

def chaveOrdTop(a):
    return (sum(a[3])/4)

def top10(bd):
    bd.sort(key=chaveOrdTop, reverse=True)
    novaBD=[]                                         
    for aluno in range(0,10):                          
        novoAluno=[]                             
        for parametro in range(0,3):                  
            novoAluno.append(bd[aluno][parametro])  
        novoAluno.append(str(sum(bd[aluno][3])/4))
        novaBD.append(novoAluno)
    return novaBD

def distribPorCurso(bd):
    distribuicao={}
    for a in bd:
        if a[2] in distribuicao.keys():
            distribuicao[a[2]]= distribuicao[a[2]]+1
        else:
            distribuicao[a[2]]= 1
    return distribuicao

def distribPorMedia(bd):
    distribuicao={}
    for a in bd:
        if round(sum(a[3])/4) in distribuicao.keys():
            distribuicao[round(sum(a[3])/4)]= distribuicao[round(sum(a[3])/4)]+1
        else:
            distribuicao[round(sum(a[3])/4)]= 1
    return distribuicao

def plotDistribPorCurso(bd):
    d=distribPorCurso(bd)
    x=d.keys()
    y=[]
    for i in x:
        y.append(d[i])
    plt.bar(x,y, color="purple")
    plt.xlabel("Cursos")
    plt.ylabel("Número de alunos")
    plt.show()
    
def plotDistribPorMedia(bd):
    d=distribPorMedia(bd)
    x=d.keys()
    y=[]
    for i in x:
        y.append(d[i])
    plt.bar(x,y, color="purple")
    plt.xlabel("Médias Arredondadas")
    plt.ylabel("Número de alunos")
    plt.show()

BD = []
BD = lerDataset("../Algoritmos/alunos.csv") 
print(BD)
listarDataset(BD)
consultarDataset(BD, "a53")
for linha in top10(BD):
    print(linha)
d=distribPorCurso(BD)
print(d)
m=distribPorMedia(BD)
print(m)
plotDistribPorCurso(BD)
plotDistribPorMedia(BD)
 
```
